```python
# 添加了所有的参数  10*3*6         输出全部公式
import time
start_time=time.time()
from gplearn.genetic import SymbolicRegressor
from sklearn.metrics import mean_absolute_error
# 设置符号回归模型
function_set = ['add', 'sub', 'mul', 'div', 'sqrt', 'log', 'abs', 'neg', 'inv']
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=22)
X_train_scaled=X_train.values
X_test_scaled=X_test.values
# feature_names = x_choose.columns.tolist()
# 生成参数网格  添加了p_subtree_mutation参数后的
parsimony_coefficient_values=np.arange(0.0005,0.0015,0.0005)
p_crossover_values=np.arange(0.5,0.95,0.05)
param_grid = {
    'population_size': [5000],
    'generations': [20],
    'p_crossover': p_crossover_values,
    'parsimony_coefficient': parsimony_coefficient_values,
    'tournament_size': [20],
    'metric': ['mean absolute error'],
    'const_range': [(-1, 1)]
}
# 存储所有结果的列表
result_train = []
result_test=[]
def perform_symbolic_regression(params):
    population_size = params['population_size']
    generations = params['generations']
    p_crossover = params['p_crossover']
    p_subtree_mutation = params['p_subtree_mutation']
    p_point_mutation = p_subtree_mutation  # 设置 p_point_mutation
    p_hoist_mutation = 1 - p_crossover - p_subtree_mutation - p_point_mutation  # 计算 p_hoist_mutation
    parsimony_coefficient = params['parsimony_coefficient']
    tournament_size = params['tournament_size']
    metric = params['metric']
    const_range = params['const_range']
    
    sr = SymbolicRegressor(
        population_size=population_size,
        generations=generations,
        p_crossover=p_crossover,
        p_subtree_mutation=p_subtree_mutation,
        p_point_mutation=p_point_mutation,
        p_hoist_mutation=p_hoist_mutation,  # 添加 p_hoist_mutation
        function_set=function_set, 
        stopping_criteria=0.01,
        parsimony_coefficient=parsimony_coefficient,
        tournament_size=tournament_size,
        metric=metric,
        const_range=const_range,
        random_state=22,
        # feature_names=feature_names,
        verbose=0,
        n_jobs=-1
    )
    sr.fit(X_train_scaled, y_train)
    # 输出符号回归模型生成的所有数学表达式
    if sr._programs is not None:
        for generation, programs in enumerate(sr._programs):
            if programs is not None:
                for program in programs:
                    if program is not None:
                        y_pred_train = program.execute(X_train_scaled)              # 对训练集求mae
                        mae_train = mean_absolute_error(y_train, y_pred_train)
                        complexity = program.length_
                        result_train.append({
                            'params': params,
                            'expression': str(program),
                            'mae': mae_train,
                            'complexity': complexity
                        })
                        
                        y_pred_test = program.execute(X_test_scaled)              # 对训练集求mae
                        mae_test = mean_absolute_error(y_test, y_pred_test)
                        complexity = program.length_
                        result_test.append({
                            'params': params,
                            'expression': str(program),
                            'mae': mae_test,
                            'complexity': complexity
                        })
# 遍历参数网格中的所有组合                        
for population_size in param_grid['population_size']:
    for generations in param_grid['generations']:
        for p_crossover in param_grid['p_crossover']:
            for p_subtree_mutation in np.arange((0.9-p_crossover)/2,(1-p_crossover)/2,0.01):
                # 计算 p_point_mutation 和 p_hoist_mutation
                p_point_mutation = p_subtree_mutation  # 设置 p_point_mutation
                p_hoist_mutation = 1 - p_crossover - p_subtree_mutation - p_point_mutation  # 计算 p_hoist_mutation
                
                for parsimony_coefficient in param_grid['parsimony_coefficient']:
                    for tournament_size in param_grid['tournament_size']:
                        for metric in param_grid['metric']:
                            for const_range in param_grid['const_range']:
                                params = {
                                    'population_size': population_size,
                                    'generations': generations,
                                    'p_crossover': p_crossover,
                                    'parsimony_coefficient': parsimony_coefficient,
                                    'p_subtree_mutation': p_subtree_mutation,
                                    'p_point_mutation': p_point_mutation,  # 添加 p_point_mutation
                                    'p_hoist_mutation': p_hoist_mutation,  # 添加 p_hoist_mutation
                                    'tournament_size': tournament_size,
                                    'metric': metric,
                                    'const_range': const_range
                                }
                                perform_symbolic_regression(params)
# 将结果存储在 DataFrame 中
train_df = pd.DataFrame(result_train)
test_df = pd.DataFrame(result_test)
# 找到每个复杂度下的最优 MAE
optimal_points_train = train_df.loc[train_df.groupby('complexity')['mae'].idxmin()]
# optimal_points_test =test_df.loc[test_df.groupby('complexity')['mae'].idxmin()]
# 找到 Pareto 前沿
def pareto_frontier(data, x_col, y_col):
    sorted_data = data.sort_values(by=[x_col, y_col], ascending=[True, True])
    pareto_front = [sorted_data.iloc[0]]
    
    for i in range(1, len(sorted_data)):
        if sorted_data.iloc[i][y_col] < pareto_front[-1][y_col]:
            pareto_front.append(sorted_data.iloc[i])
    
    return pd.DataFrame(pareto_front)
pareto_df_train = pareto_frontier(optimal_points_train, 'complexity', 'mae')
# 根据训练集中的最优公式输出这些公式在测试集中的 MAE，然后输出
optimal_test_results = test_df[test_df['expression'].isin(pareto_df_train['expression'])]
# 去除测试集中的重复项，保留最优MAE值
optimal_test_results = optimal_test_results.loc[optimal_test_results.groupby('expression')['mae'].idxmin()]
# 保存结果为 xlsx文件
pareto_df_train.to_excel('train.xlsx', index=False)
optimal_test_results.to_excel('test.xlsx', index=False)
end_time=time.time()
tol_time=end_time-start_time
print(tol_time)
```
